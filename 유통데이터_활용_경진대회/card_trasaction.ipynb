```python
def categorize_age(age):
    if age['age_group']>= 20 and  age['age_group'] <= 29: 
        return '20~29'
    elif  age['age_group'] >= 30 and  age['age_group']<=39: 
        return '30~39'
    elif  age['age_group']>= 40 and  age['age_group'] <=49: 
        return '40~49'
    elif  age['age_group'] >= 50 and  age['age_group'] <= 59: 
        return '50~59'
    elif  age['age_group'] >= 60 and  age['age_group'] <= 69: 
        return '60~69'
    else: 
        return '70+'
```

```python
card['agedivided'] = card.apply(categorize_age,axis = 1)
```

# 2022년,2023년 나누기

```python
card.head(4)
```

```python
card['pay_day_of_week'] = card['pay_day_of_week'].astype(str)
card.info()
```

```python
card['Year'] = card['pay_date'].dt.year 
card.head(5)
```

```python
#2022년 

data1 = card[card['Year'] == 2022]
data2 = card[card['Year'] == 2023]
```

```python
data1['pay_day_of_week'].value_counts()
```

```python
data2['pay_day_of_week'].value_counts()
```

```python
data1['pay_day_of_week'] = data1['pay_day_of_week'].replace({'1':'일요일','2':'월요일','3':'화요일','4':'수요일','5':'목요일','6':'금요일','7':'토요일'})
```

```python
data2['pay_day_of_week'] = data2['pay_day_of_week'].replace({'1':'일요일','2':'월요일','3':'화요일','4':'수요일','5':'목요일','6':'금요일','7':'토요일'})
```

```python
data1_group = data1.groupby('pay_day_of_week')['amount'].sum()
data2_group = data2.groupby('pay_day_of_week')['amount'].sum()
```

```python
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
#2022년 
plt.figure(figsize = (15,10))
plt.bar(data1_group.index, data1_group.values, color = 'orange')

total_cost = data1_group.sum()


for i, v in enumerate(data1_group.values):
    percentage = (v / total_cost) * 100
    plt.text(i, v + 10, f'{percentage:.2f}%', ha='center', va='bottom', fontsize=14)
    
plt.xlabel('요일', fontsize = 20)
plt.ylabel('총 비용의 백분율',fontsize = 20)
plt.title('요일별 카드 사용금액 비율(2022년)',fontsize = 20)
plt.xticks(fontsize=20,rotation=45)                  # X-axis tick labels
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```

```python
#2023년 
plt.figure(figsize = (15,10))
plt.bar(data2_group.index, data2_group.values, color = 'green')

total_cost = data1_group.sum()


for i, v in enumerate(data2_group.values):
    percentage = (v / total_cost) * 100
    plt.text(i, v + 10, f'{percentage:.2f}%', ha='center', va='bottom', fontsize=14)
    
plt.xlabel('요일', fontsize = 20)
plt.ylabel('총 비용의 백분율',fontsize = 20)
plt.title('요일별 카드 사용금액 비율(2023년)',fontsize = 20)
plt.xticks(fontsize=20,rotation = 45)                  # X-axis tick labels
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```
