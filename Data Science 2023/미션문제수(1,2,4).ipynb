## Q2

```python
import pandas as pd 

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df.head(5)
```

```python
df['Country'].value_counts()
```

```python
df1 = df.groupby(['Year', 'Country']).agg({'Spending_USD': 'sum', 'Life_Expectancy': 'mean'}).reset_index()
df1
```

```python
pivot_table = pd.pivot_table(df1, values='Life_Expectancy', index=['Year'], columns='Country',aggfunc="mean", fill_value=0).reset_index().rename_axis(None,axis =1)
pivot_table
```

```python
df2 = pivot_table.sort_values(by = "Year", ascending= False)
df2
```
